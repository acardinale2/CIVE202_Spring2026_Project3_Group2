# Part 4 - Simulating Specific Trajectory

In this section we will choose a specific trajectory to simulate vehicle behavior. 

---

## 4.1 Define the IDM Acceleration Function

This function will take the current state of the follower vehicle and the leader vehicle, along with the model parameters, and returns the acceleration.

The function is based on the IDM Equations. It uses information from the leader vehicle and follower vehicle to output acceleration and deceleration.

IDM Equation:
    
* s_star = s0 + v * T + (v * delta_v) / (2 * np.sqrt(a * b))
    
where 
v: current speed of the follower vehicle (m/s)
s: current gap distance between leader and follower (m)
delta_v: speed difference = follower speed - leader speed (m/s)
v0: desired velocity (m/s)
s0: minimum spacing (m)
T: desired time headway (s)
a: maximum acceleration (m/s^2)
b: comfortable deceleration (m/s^2)
delta: acceleration exponent

```python
#Defines the IDM acceleration function
def idm_acceleration(v, s, delta_v, v0, s0, T, a, b, delta):
    #calculates the desired minimum gap s_star
    s_star = s0 + v * T + (v * delta_v) / (2 * np.sqrt(a * b))
    
    #this ensures that s_star is not negative
    s_star = max(s_star, s0)
    
    #This calculates the acceleration using the IDM equation
    acceleration = a * (1 - (v/v0)**delta - (s_star/s)**2)
    
    return acceleration
```

## 4.2 Run Function with Values

Now, we can run the function with the typical values to get the acceleration.
    
Typical IDM values are:
v0 = 30 (m/s), desired velocity
s0 = 2 (m), minimum spacing
T = 1.5 (s), desired time headway
a = 1.0 (m/s^2), maximum acceleration
b = 1.5 (m/s^2), comfortable deceleration

Test values can be adjusted to test different cases.

The test variables are v (m/s), s (m), and delta_v (m/s).


```python
#define parameters
v0 = 30    #desired velocity (m/s)
s0 = 2     #minimum spacing (m)
T = 1.5    #desired time headway (s)
a = 1.0    #maximum acceleration (m/s^2)
b = 1.5    #comfortable deceleration (m/s^2)
delta = 4  #acceleration exponent

# Test case
test_v = 20        # follower speed
test_s = 25        # gap distance
test_delta_v = 2   # follower is faster than leader by 2 m/s

# Call the function and print the result
test_acc = idm_acceleration(test_v, test_s, test_delta_v, v0, s0, T, a, b, delta)
print('test acceleration =', test_acc, '(m/s^2)')
```

## 4.3 Set Up Simulation

With the IDM function working, we can now simulate a vehicle following scenario over time. 
We will utilize Euler's method to do this.

The at every small step the acceleration will be calculated from IDM according to teh dt, or change in time. 

For each step the parameters will be used in the following way:
* New acceleration = IDM function output
* New speed = old speed + acceleration × dt
* New position = old position + speed × dt

We will simulate a follower vehicle tracking a leader vehicle. 

```python
#selecting a trajectory pair
trajectory_number = 1

#Selecting the data based on the subset
data_subset = ngsim_data[ngsim_data['trajectory_number'] == trajectory_number].reset_index(drop=True)

#extracting the leader's position and speed over time
leader_position = data_subset['leader_position(m)'].values
leader_speed = data_subset['leader_speed(m/s)'].values
leader_acc = data_subset['leader_acc(m/s^2)'].values
time = data_subset['Time'].values

#Identifying the time step in our data
dt = 0.1 #means every 0.1 second the acceleration will be calulated
```

## 4.4 Run the IDM Simulation

Now we can run the simulation. 

```python
#Set the IDM parameters
v0 = 30    #desired velocity (m/s)
s0 = 2     #minimum spacing (m)
T = 1.5    #desired time headway (s)
a_param = 1.0    #maximum acceleration (m/s^2)  
b = 1.5    #comfortable deceleration (m/s^2)
delta = 4  #acceleration exponent

#Initialize arrays to store simulation results
n_steps = len(time)                      #number of simulation steps we will take
sim_position = np.zeros(n_steps)         #simulated follower position
sim_speed = np.zeros(n_steps)            #simulated follower speed
sim_acc = np.zeros(n_steps)

#created zero vectors to hold storage for the values
#Now we fill those vectors with the starting value we know of
sim_position[0] = data_subset['follower_position(m)'].values[0]
sim_speed[0] = data_subset['follower_speed(m/s)'].values[0]

#Run the simulation using a for loop
for i in range(n_steps - 1):
    #Calculate the gap and speed difference at the time step
    gap = max(leader_position[i] - sim_position[i], 0.1)
    delta_v = sim_speed[i] - leader_speed[i]
    
    #Use the IDM function to calculate acceleration
    sim_acc[i] = idm_acceleration(sim_speed[i], gap, delta_v, v0, s0, T, a_param, b, delta)

    #Update speed and position for the next time step
    sim_speed[i+1] = max(sim_speed[i] + sim_acc[i]*dt, 0) #updates speed for next iteration
    sim_position[i+1] = sim_position[i] +sim_speed[i]*dt
    
#if successuly completed, will print simulation complete
print("Simulation complete!")
```

## 4.5 Plot the Simulation Results

We will now plot our results so that we can compare it against the NGSIM data. 

```python
#creating acceleration plot
plt.figure(figsize=(12,6))

plt.plot(time, leader_acc, 'k--', linewidth = 2, label = 'Leader (NGSIM)')
plt.plot(time, data_subset['follower_acc(m/s^2)'].values, 'b-', linewidth = 2, label = 'Real Follower (NGSIM)')
plt.plot(time, sim_acc, 'r-', linewidth = 2, label = 'Simulated Follower (NGSIM)')

#customization
plt.xlabel("Time (seconds)")
plt.ylabel("Acceleration (m/s^2)")
plt.title("Acceleration of Leader, Real Follower, and Simulated Follower")
plt.legend()
plt.grid(True)
plt.savefig("project3_SIM_acc.png", dpi = 300, bbox_inches='tight')
plt.show()
```
